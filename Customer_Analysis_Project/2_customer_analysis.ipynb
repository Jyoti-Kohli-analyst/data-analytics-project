```python
sns.barplot(data = df, x = 'Genre', y = 'Spending Score (1-100)')
plt.title ("Average Spending Score by Gender")
plt.show()
            
```

```python
sns.histplot(df['Age'], bins = 10)
plt.title ("Customer Age Distribution")
plt.show()
```

```python
sns.scatterplot(data = df, x = 'Age', y = 'Spending Score (1-100)')
plt.title ("Age vs Spending score")
plt.show()
```

```python
sns.scatterplot(
    data = df, 
    x = 'Annual Income (k$)',
    y = 'Spending Score (1-100)',
    hue = 'Genre'
)
plt.title("Income vs Spending score")
plt.show()
```

```python
high_income_low_spend = df[
(df['Annual Income (k$)'] > df['Annual Income (k$)'].mean()) &
    (df['Spending Score (1-100)'] < df['Spending Score (1-100)'].mean())
    ]
high_income_low_spend.head()
```

```python
len(high_income_low_spend)
```

```python
high_value = df[
(df['Annual Income (k$)'] > df['Annual Income (k$)'].mean()) &
    (df['Spending Score (1-100)'] > df['Spending Score (1-100)'].mean())
    ]
high_value.head()
```

```python
len(high_value)
```
